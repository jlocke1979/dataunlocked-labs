## Focused Analysis (2025)

To align with the visualization, the dataset was filtered to a single year (2025). 

This removes duplication across years and allows direct comparison across organ categories. The resulting data shows the relative scale of transplant activity by organ type, with kidney transplants dominating overall volume.

```python
df1_2025 = clean1[

    (clean1["year"] == 2025) &

    (clean1["donor_type"] == "All Donor Types")

].copy()

df1_2025 = df1_2025[

    (~df1_2025["organ"].str.contains("All")) &

    (~df1_2025["organ"].str.contains("VCA"))

]

df1_2025 = df1_2025[["organ", "transplants"]]

df1_2025 = df1_2025.sort_values("transplants", ascending=False).reset_index(drop=True)

df1_2025
```
