## 8. Design Decisions

- **No flow lines in final view** — Volume bubbles on fixed positions from geocoded nodes; edges used only to build organ-specific totals.
- **Global sqrt scale (unified)** — Avoids misleading “every organ looks equally peaked” when scales reset per view.
- **Transplant default** — Audience cares most about where transplants occur; donor view is one click away.
- **Organ colors on map + picker only** — Legend labels stay charcoal for readability; color signals active organ filter.
- **Partial coordinates** — Collapsible audit notes in the app: **192 of 246** transplant centers plotted (**38,044 / 41,808** transplants, ~**91%** of 2025 all-organ summary volume); 54 centers lack lat/lon. Geocode mix: manual (highest trust), Nominatim/facility, state-centroid fallback (notably some CA university centers).
